```python
# Creating a duplicate of original dataset before make any changes in it
hb_df1 = hb_df.copy()
```

# What is hotel booking cancellation rate ?

```python
# Creating a pia chart
cancel_counts = hb_df1['is_canceled'].value_counts()
print(f'total bookings : {cancel_counts.sum()}')
print(f'canceled booking : {cancel_counts[1]}')
# Create labels and colors
labels = ['Not Canceled', 'Canceled']
colors = ['lightgreen', 'coral']

# Plot pie chart
plt.figure(figsize=(6,6))
plt.pie(cancel_counts, labels= labels, explode=(0.05,0.05), autopct='%1.2f%%', startangle=90, colors=colors)
plt.title('Booking Cancellation Rate')
plt.axis('equal') # Equal aspect ratio to make pie circular
plt.show()
```

Insights from the Booking Cancellation Rate Pie Chart

The pie chart shows that approximately 27.5% of all bookings were canceled, while 72.5% were honored. This highlights a relatively high cancellation rate, particularly concerning for hotel operations. Strategies to reduce cancellations and improve booking reliability may be necessary, especially for bookings made through online travel agents or with long lead times.

# Which hotel makes more revenue


```python
# Calculat total revenue for each booking
hb_df1['revenue'] = hb_df1['adr'] * hb_df1['total_stay']
total_revenue = hb_df1.groupby('hotel')['revenue'].sum()
print(total_revenue)

#ploting the value in pie chart
plt.figure(figsize=(8,6))
plt.pie(total_revenue, labels= total_revenue.index, explode=(0.02,0.02), autopct='%1.2f%%', startangle=140)
plt.title('Total Revenue by Hotel')
plt.show()
```

# Insight from above chart:
The pie chart shows that City Hotel contributes a larger share of the total revenue, confirming it earns more overall than Resort Hotel. This is due to higher booking volume, even though Resort Hotel guests usually stay longer

# Which hotel has the longer waiting time

```python
waiting_time = hb_df1.groupby('hotel')['days_in_waiting_list'].mean().sort_values()
print('Average waiting time for each hotel: ')
print(waiting_time)
# Ploting bar chart
plt.figure(figsize=(6,4))
sns.barplot(x=waiting_time.index, y=waiting_time.values)
plt.title('Average Waiting Time for Each Hotel')
plt.xlabel('Hotel')
plt.ylabel('Average Waiting Time (Days)')
plt.show()
```

# Insights from above chart
The bar plot shows that City Hotel has a longer waiting time before guests arrive. This means people usually book City Hotel earlier in advance. On the other hand, Resort Hotel is booked closer to the travel date, likely for quick or last-minute vacation plans.

# What is adr across different months

```python
monthly_adr = hb_df1.groupby(['arrival_date_month', 'hotel'])['adr'].mean().reset_index()
# Creating a month list
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
monthly_adr['arrival_date_month'] = pd.Categorical(monthly_adr['arrival_date_month'], categories= months, ordered=True)
print(monthly_adr)

monthly_adr = monthly_adr.sort_values('arrival_date_month')
print(monthly_adr)
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x= monthly_adr['arrival_date_month'], y= monthly_adr['adr'], hue = monthly_adr['hotel'])
plt.title('Average Daily Rate across different months')
plt.xlabel('Month')
plt.ylabel('Average Daily Rate')
plt.show()
```

# Insights from above chart
The above plot shows that ADR is higher during summer months like July and August, indicating peak season when hotel prices rise due to high demand. In contrast, ADR is lower in months like January and November, showing off-season periods with cheaper rates. This pattern reflects seasonal pricing in the hotel industry
